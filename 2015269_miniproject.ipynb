# ML For Prediction

```python
import pandas as pd
import numpy as np  
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
import math
from sklearn.model_selection import cross_val_score
```

```python
dataset_symptoms=pd.read_csv('F:\\dataset_mini\\dataset.csv')
dataset_Symptom_severity=pd.read_csv('F:\\dataset_mini\\Symptom_severity.csv')
```

```python
dataset_symptoms.sample(frac=1)#shuffule data and return fraction of data u want
```

```python
#pd.set_option('display.max_columns', None) # none is no of of columns u want 2 see
#pd.set_option('display.max_rows', None)
```

```python
dataset_Symptom_severity.head()
```

```python
dataset_symptoms=dataset_symptoms.fillna(0)
```

```python
dataset_symptoms
```

```python
y=dataset_symptoms['Disease']
dataset_symptoms=dataset_symptoms.iloc[:,1:]
```

```python
y.unique()#42 unique disease
```

```python
dataset_symptoms.head()
```

```python
y.head()
```

```python
key=[]
value=[]
[key.append(i) for i in dataset_Symptom_severity['Symptom']]
[value.append(i) for i in dataset_Symptom_severity['weight']]
```

```python
dataset_symptoms=dataset_symptoms.replace(key,value,regex=True)#without regrex does not replace all elements
dataset_symptoms=dataset_symptoms.replace('dischromic _patches','0',regex=True)
dataset_symptoms=dataset_symptoms.replace('spotting_ urination','0',regex=True)
dataset_symptoms=dataset_symptoms.replace('foul_smell_of urine','0',regex=True)
```

```python
dataset_symptoms.head()
```

```python
# none is no of of columns u want 2 see
#pd.set_option('display.max_rows', None)
```

```python
dataset_symptoms
```

```python
for x in dataset_symptoms.columns:
    print(x ,':', len(dataset_symptoms[x].unique()))
```

```python

X_train, X_test, Y_train, Y_test = train_test_split(dataset_symptoms, y, test_size=0.4, random_state=1)# 0.4 use in test random state is same every time(if no.) but by default is none
```

```python
gini_model = DecisionTreeClassifier(criterion = "gini",
            random_state = 100,max_depth=3, min_samples_leaf=5)
gini_model.fit(X_train, Y_train)
```

```python
y_pred = gini_model.predict(X_test)
accuracy_score(Y_test,y_pred)*100
```

```python
entropy_model = DecisionTreeClassifier(criterion = "entropy",
            random_state = 100,max_depth=3, min_samples_leaf=5)
entropy_model.fit(X_train, Y_train)
```

```python
y_pred = entropy_model.predict(X_test)
accuracy_score(Y_test,y_pred)*100
```

```python
random_forest_classifier = RandomForestClassifier(n_estimators = 1400,max_depth=30,max_features='sqrt',min_samples_split=5)# higher than depth increase chances of overfitting
random_forest_classifier.fit(X_train, Y_train)
```

```python
y_pred = random_forest_classifier.predict(X_test)
accuracy_score(Y_test,y_pred)*100
```

```python
print(classification_report(Y_test, y_pred))
print(confusion_matrix(Y_test, y_pred))
```

```python
Knn_classifier = KNeighborsClassifier(n_neighbors=5)
Knn_classifier.fit(X_train, Y_train)
```

```python
y_pred = Knn_classifier.predict(X_test)
accuracy_score(Y_test,y_pred)*100
```
