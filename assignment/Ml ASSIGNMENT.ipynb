```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.cluster.hierarchy as shc
import sklearn.metrics as sm
from sklearn.cluster import AgglomerativeClustering
```

```python
dataset=pd.read_csv('./customers.csv')
subset=dataset.iloc[:,[3,4]].values
```

```python
D=shc.dendrogram(shc.linkage(subset,method='ward'))
plt.title("Dendrogram")
plt.axhline(130)
plt.show()
```

```python
clustering_obj=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
clustering_obj.fit_predict(subset)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(subset[:,0],subset[:,1],c=clustering_obj.labels_, cmap='rainbow')
```
